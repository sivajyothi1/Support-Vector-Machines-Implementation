```python
df.columns
```

```python
## independent and dependent features
X = df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y = df['total_bill']
```

```python
## Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=42)
```

```python
X_train
```

```python
X_train.head()
```

```python
## Feature Encoding(label encoding and onehot encoding)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
X_train['sex'] = le1.fit_transform(X_train['sex'])
X_train['smoker'] = le2.fit_transform(X_train['smoker'])
X_train['time'] = le3.fit_transform(X_train['time'])
```

```python
X_train.head()
```

```python
X_test['sex'] = le1.transform(X_test['sex'])
X_test['smoker'] = le2.transform(X_test['smoker'])
X_test['time'] = le3.transform(X_test['time'])
```

```python
X_test.head()
```

```python
## Onehot encoding --- columnTransformer

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers = [('onehot', OneHotEncoder(drop='first'),[3])],
                         remainder='passthrough')
```

```python
import sys
import numpy as np
np.set_printoptions(threshold = sys.maxsize)
X_train = ct.fit_transform(X_train)
```

```python
X_test = ct.transform(X_test)
```

```python
X_test
```

```python
### SVR -- Support Vector Regression
from sklearn.svm import SVR
svr = SVR()
```

```python
svr.fit(X_train, y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
      
```

```python
## Hyperparameter tuning using GridSearch CV
```

```python
from sklearn.model_selection import GridSearchCV

## Defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.001],
              'kernel':['rbf']}
```

```python
grid = GridSearchCV(SVR(), param_grid, refit= True, verbose=3)

# fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid_prediction = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
print(r2_score(y_test, grid_prediction))
print(mean_absolute_error(y_test, grid_prediction))
      
```
